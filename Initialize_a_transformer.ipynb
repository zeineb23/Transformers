Pretrained transformer model from checkpoint = config file (plan to construct the model from its architecture with random weights) + model file (the actual weights)

We can create our own trained transformer model based on Bert's same architecture with random weights.

```python
from transformers import BertConfig, BertModel

# Building the config
config = BertConfig()

# Building the model from the config
model = BertModel(config)
```

```python
print(config) # Model is randomly initialized! and not trained!
```
